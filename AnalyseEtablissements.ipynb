```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime, os
```

```python
df_files = os.listdir('../../data/DataGithub/')
df_files
```

```python
df = pd.read_csv('244400404_effectifs-eleves-cleaned')
df['datetime']= df['Date'].apply( lambda string1 : datetime.datetime.strptime(string1,"%Y-%m-%d"))
df['joursem']= df['datetime'].apply( lambda date1 : date1.weekday())
df['isoweek']= df['datetime'].apply( lambda date1 : date1.isocalendar().week)
df['AnScol']= df['datetime'].apply(lambda date1 : date1.year if date1.isocalendar().week>30 else date1.year-1)
df.head()
```

```python
df[['AnScol','datetime']].groupby(by='datetime').median().reset_index().groupby(by='AnScol').count()
```

```python
df[['An','datetime']].groupby(by='datetime').median().reset_index().groupby(by='An').count()
```

```python
df_ets22=df[df['An']==2022][['Nom du site','Réel']].groupby(['Nom du site']).median()
df_ets22=pd.DataFrame(df_ets22).reset_index()
df_ets22.rename(columns={'Nom du site':'Ecole'}, inplace=True)
df_ets22.sort_values(by='Ecole').head(30)
```

```python
setEts22=df_ets22['Ecole'].unique()
len(setEts22)
```

```python
df[df['Nom du site'].isin(setEts22)][['Nom du site','AnScol','datetime']].groupby(by=['Nom du site','AnScol']).\
count().reset_index().pivot(values='datetime', index=['Nom du site'],columns=['AnScol']).\
count(axis=1).sort_values().head(30)
```
